### Perform actions on a report
Set a custom theme for a report based on a Power BI theme file

```python
#Sets a custom theme for a report based on a theme .json file.
#Parameters:theme_file_path (str) – The file path of the theme.json file. This can either be from a Fabric lakehouse or from the web. Example for lakehouse: file_path = ‘/lakehouse/default/Files/CY23SU09.json’ 
#Example for web url: file_path = ‘https://raw.githubusercontent.com/PowerBiDevCamp/FabricUserApiDemo/main/FabricUserApiDemo/DefinitionTemplates/Shared/Reports/StaticResources/SharedResources/BaseThemes/CY23SU08.json’

rpt.set_theme(theme_file_path='/lakehouse/default/Files/CY23SU09.json')
```

```python
rpt.set_theme(theme_file_path='https://raw.githubusercontent.com/PowerBiDevCamp/FabricUserApiDemo/main/FabricUserApiDemo/DefinitionTemplates/Shared/Reports/StaticResources/SharedResources/BaseThemes/CY23SU08.json')
```

Set whether a Power BI report page is hidden or visible

```python
#Sets whether a report page is visible or hidden.
rpt.set_page_visibility(page_name='Broken Visuals', hidden=True)
```

Set the active page (page shown upon opening the report)

```python
#Sets the active page (first page displayed when opening a report) for a report.
rpt.set_active_page(page_name='Reset slicers')
```

```python
labs.report.launch_report(report='Semantic Link Labs', workspace=None)
```

Disable the 'show items with no data' for all visuals within a Power BI report

```python
#Disables the show items with no data property in all visuals within the report
rpt.disable_show_items_with_no_data()
```

Remove all custom visuals in a Power BI report which are not used in any visuals within the report

```python
#Removes any custom visuals within the report that are not used in the report.
rpt.remove_unnecessary_custom_visuals()
```

Hide all pages which are used for a tooltip or drillthrough

```python
#Hides all tooltip pages and drillthrough pages in a report.
rpt.hide_tooltip_drillthrough_pages()
```
